```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = "https://www.thegioididong.com/dtdd/samsung-galaxy-s22-ultra?code=0131491002959"
res = requests.get(url, headers={
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36"
    })
soup = BeautifulSoup(res.content, features="lxml")
print(soup.prettify())
```

```python
product = {}
```

```python
name = soup.find("h1")
product["name"] = name.string 
print(name)
print(product["name"])
```

```python
color = soup.find(name ="a", attrs={"data-color": True, "class": "box03__item item act"})
product["color"] = color.string
print(color)
print(product["color"])
```

```python
model = soup.find(name= "a", attrs={"data-index": True, "class": "box03__item item act"}, href=re.compile("^/dtdd"))
product["model"] = model.string
print(model)
print(product["model"])
```

```python
price = soup.find(name= "p", attrs={"class": "box-price-present"})
product["price"] = price.string
print(price)
print(product["price"])
```

```python
parameters = soup.select_one("ul.parameter__list").find_all("li")

for parameter in parameters:
    key = parameter.select_one("p.lileft").string
    value = ""
    for string in parameter.select_one("div.liright").stripped_strings:
        value = value + string + ", "
    value = value[:-2]
    product[key] = value
    print("%s %s" % (key,value))
```

```python
print(product)
```
